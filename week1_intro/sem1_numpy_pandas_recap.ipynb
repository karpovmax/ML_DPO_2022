# Генерация фичей

```python
data['FamilySize'] = data['SibSp'] + data['Parch']
data
```

## Генерация фичей с lambda x в методе apply

```python
data['Alone'] = data.FamilySize.apply(lambda x: 1 if x == 0 else 0)
data.head()
```

```python
data.loc[0]['Name']
```

```python
data.iloc[0]['Name'].split(',')[1].split('.')[0].strip()
```

```python
def return_title(full_name):
    return(full_name.split(',')[1].split('.')[0].strip())
```

```python
data['Title'] = data['Name'].apply(return_title)
```

```python
data
```

```python
data['Title'].value_counts()
```

```python
data['Title'] = data['Title'].apply(lambda x: 'Misc' if x not in \
                                   ['Mr', 'Miss', 'Mrs', 'Master', 'Dr'] else x)
```

```python
data['Title'].value_counts()
```

```python
data = pd.get_dummies(data, columns=['Title'])
```
